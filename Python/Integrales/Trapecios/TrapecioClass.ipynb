```python
import sympy as sp
import math

import sympy as sp

class IntegralT:

    def __init__(self, a = None, b = None, n = 999) -> None:
        self.n = n
        self.f = None
        self.a = a
        self.b = b
        self.deltax = None
        self.lista_xi = []
        self.lista_funciones = []
        self.resultado = None
        self.x = sp.Symbol('x')

    def set_f(self, funcion):
        self.f = sp.sympify(funcion)
        
    def get_f(self):
        return self.f
    
    def set_n(self, n):
        self.n = n
    
    def set_intervalo(self, a, b):
        self.a = a
        self.b = b
    
    def generar_variables(self):
        self.deltax = (self.b - self.a) / self.n
    
    def generar_resultado(self):
        self.generar_variables()
        for i in range(self.n + 1):
            if (i in [0, self.n]):
                xi = self.a + (i * self.deltax)
                self.lista_xi.append(xi)
                fxi = self.f.subs(sp.Symbol('x'), xi)
                self.lista_funciones.append(fxi)
            else:
                xi = self.a + (i * self.deltax)
                self.lista_xi.append(xi)
                fxi = self.f.subs(sp.Symbol('x'), xi) * 2
                self.lista_funciones.append(fxi)
            
        self.resultado = ((self.deltax) / 2) * sum(self.lista_funciones)
        
    def __repr__(self):
        
        return f"""[-* Datos integral *-]
n = {self.n}
intervalo = [{self.a},{self.b}]
Δx = {self.deltax}
resultado = {self.resultado}
"""
    def ver(self):
        print(self)


        
# Constantes matemáticas
e = math.e
pi = math.pi

# Para ignorar el warning
x = sp.Symbol('x')

# 1) e^(x^2)
f1 = IntegralT(0, 3)
f1.set_f(e**(x**2))
f1.generar_resultado()
f1.ver()

# 2) (x)(sin(x))^(1/2)
f2 = IntegralT(0, pi)
f2.set_f(x*sp.sin(x)**(1/2))
f2.generar_resultado()
f2.ver()

# 3) (e^2x)/(3x)
f2 = IntegralT(1,2)
f2.set_f((e**2*x)/(3*x))
f2.generar_resultado()
f2.ver()
```
